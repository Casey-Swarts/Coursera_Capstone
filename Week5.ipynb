Introduction/Business Problem:

For this project I wanted to focus on my current city and drill down to the local beach to find a good location to open a coffee shop. I'm wanting to find a location that can be accessible from the beach and possibly is in close proximity to a bakery or pastry shop. This type of research would be an interest to any entrepreneur hoping to move into owning a coffee franchise/business.   

Data:
    
To solve this business problem, I will collect location data from Foursquare and apply the techniques learned in previous lessons. The data collected will involve comparing the locations of coffee shops and bakeries in relation to Moonlight Beach. By examining the locations we can find the best area to start a coffee shop.

Methodology:

The data was accessed through the FourSquare API interface. The data was then visualized using folium package to see the number of coffee shops and bakerys near Moonlight Beach.

Results/Discussion:

From the comparison rendered we can see that there are a good amount of coffee shops in the area except for locations closer to the beach and farther down south. When compared over the bakery selection, which are less frequent, the southernmost section of the beach would be a great location for another coffee shop with having almost all of the bakeries nearby and less coffee shops.

Conclusion:

With visualizing the distribution of coffee shops and bakeries in the area we have found that South Encinitas nearest Moonlight Beach would make for the best location for a new coffee shop.

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']
# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

```python
dataframe.shape
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python
# assign relevant part of JSON to venues
venues = results['response']['venues']

# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

```python
## keeping only the data frames that is related to venues and locations
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
df_fil_bakery = dataframe.reindex(columns = filtered_columns)

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else: 
        return(categories_list)[0]['name']
    
# filter the category for each row
df_fil_bakery['categories'] = df_fil_bakery.apply(get_category_type, axis=1)

# clean column names by keeping only last term
df_fil_bakery.columns = [column.split('.')[-1] for column in df_fil_bakery.columns]

df_fil_bakery.head()
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start =15)

#To locate Moonlight Beach, using red mark on the map
folium.CircleMarker(
    [latitude, longitude],
    radius=20,
    color = 'red',
    popup='Moonlight Beach',
    fill = 'True',
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)

# To locate the Bakeries, using green mark on the map
for lat, lng, label in zip(df_fil_bakery.lat, df_fil_bakery.lng, df_fil_bakery.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=10,
        color = 'green',
        popup = label,
        fill = 'True'
).add_to(venues_map)

venues_map
```
